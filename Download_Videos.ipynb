```
# Download to your computer
from google.colab import files
import zipfile

with zipfile.ZipFile('asl_video_clips.zip', 'w') as z:
    for f in os.listdir('video_clips'):
        if f.endswith('.mp4'):
            z.write(f'video_clips/{f}', f)

size = os.path.getsize('asl_video_clips.zip') / 1e6
print(f'Zipped: asl_video_clips.zip ({size:.0f} MB)')
files.download('asl_video_clips.zip')
```
